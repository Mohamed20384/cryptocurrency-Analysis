```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style='whitegrid')
plt.rcParams["figure.figsize"] = (14, 6)

df = pd.read_csv('bitcoin_data.csv', parse_dates=['date'], index_col='date')
print("Data loaded successfully.")
```

```python
print("Dataset Shape:", df.shape)
display(df.head())
display(df.tail())
```

```python
missing = df.isnull().sum()
print("\nMissing Values:")
print(missing[missing > 0])
```

```python
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[column] < Q1 - 1.5 * IQR) | (data[column] > Q3 + 1.5 * IQR)]
    return outliers

outlier_cols = ['price', 'daily_return', 'volatility_30']
for col in outlier_cols:
    outliers = detect_outliers_iqr(df, col)
    print(f"\n{col} â†’ {len(outliers)} outliers")
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
df['price'].plot(title='Bitcoin Price Over Time', color='blue')
plt.ylabel('Price (USD)')
plt.show()

df['volume'].plot(title='Bitcoin Volume Over Time', color='orange')
plt.ylabel('Volume')
plt.show()

df['daily_return'].plot(title='Daily Return (%) Over Time', color='green')
plt.ylabel('Return %')
plt.show()
```

```python
for col in ['daily_return', 'volatility_30', 'momentum_7']:
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
cor_target = df.corr(numeric_only=True)['target'].sort_values(ascending=False)
print("\nCorrelation with Target (Next Day Price):")
display(cor_target)
```

```python
top_corr_features = cor_target[1:6].index 
print(top_corr_features)
sns.pairplot(df[top_corr_features.to_list() + ['target']])
plt.suptitle("Top Correlated Features with Target", y=1.02)
plt.show()
```
